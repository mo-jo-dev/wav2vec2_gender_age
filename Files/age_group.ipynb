```python
import pandas as pd
```

```python
df = pd.read_csv('output_file.csv')
# df2 = pd.read_csv('y_test.csv')
```

```python
bins = [6, 8, 13]
labels = ['6-8', '9-11']
df['Age Group'] = pd.cut(df['AGE'], bins=bins, labels=labels, right=True, include_lowest=True)
# df2['Age Group'] = pd.cut(df2['Age'], bins=bins, labels=labels, right=True, include_lowest=True)
```

```python
df.to_csv("output_age_group.csv", index=False)
# df2.to_csv("y_test_age_group.csv", index=False)
df
```
